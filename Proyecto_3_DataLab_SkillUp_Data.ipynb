##4.1.9 Crear variables nuevas (Daily Return)

El rendimiento de una acciÃ³n es el beneficio total que un inversor obtiene de su inversiÃ³n en acciones, expresado como un porcentaje del capital invertido en un periodo determinado

Creamos una columna nueva donde se exprese el **rendimiento diario** (sin considerar dividendos, que son el dinero que te pago la empresa)


**Como se calcula?**

Para calcular el rendimiento diario de una acciÃ³n, usa la fÃ³rmula:

 **((Precio de Hoy - Precio de Ayer) /
  Precio de Ayer)x100**
  
Esta fÃ³rmula darÃ¡ la variaciÃ³n porcentual entre el precio de cierre de un dÃ­a y el del dÃ­a anterior. Si la acciÃ³n paga dividendos, se pueden sumar al cÃ¡lculo del rendimiento total, pero para el rendimiento diario puro, se suele centrar solo en la variaciÃ³n del precio.

**Ejemplo:**

($120 - $110) / $110 = $10 / $110 â‰ˆ 0.0909
Multiplica por 100 para obtener el porcentaje: 0.0909 * 100 = 9.09%
Esto significa que la acciÃ³n tuvo un rendimiento diario del 9.09%.


**Consideraciones**
**Precios de cierre ajustado:**

Para un cÃ¡lculo mÃ¡s preciso en anÃ¡lisis financieros, se suelen usar los "precios de cierre ajustados" que ya tienen en cuenta eventos corporativos como los desdoblamientos de acciones.


```python

def calcular_rendimiento(df_prices_return, col_precio="adj_close", nueva_col="daily_return"):
  #The pct_change() method in Pandas calculates the percentage change between the current element and a prior element in a Series or DataFrame.
  #By default, it computes the percentage change from the immediately preceding row.
  # This method is particularly useful for analyzing time-series data to understand growth rates or relative changes over time.
    df_prices_return = df_prices_merged.copy()
    df_prices_return[nueva_col] = df_prices_return[col_precio].pct_change()*100
    return(df_prices_return)


df_prices_cleaned = calcular_rendimiento(df_prices_merged)
```

##4.2.3 Calcular medidas de tendencia central

```python
def resumen_estadistico(df, columna):
    serie = df[columna].dropna()  # eliminamos nulos para evitar errores
    Q1 = serie.quantile(0.25)
    Q3 = serie.quantile(0.75)
    IQR = Q3 - Q1
    resumen = {
        "Media": serie.mean(),
        "Mediana": serie.median(),
        "Moda": ', '.join(map(str, serie.mode().tolist())),  # puede haber varias modas
        "MÃ­nimo": serie.min(),
        "MÃ¡ximo": serie.max(),
        "Rango": serie.max() - serie.min(),
        "Varianza": serie.var(),  # muestral
        "DesviaciÃ³n estÃ¡ndar": serie.std(),
        "IQR": IQR,
        "Coeficiente de variaciÃ³n": serie.std() / serie.mean() if serie.mean() != 0 else np.nan
    }
    return pd.DataFrame(resumen, index=[columna])

#El IQR (Rango IntercuartÃ­lico) es una medida de dispersiÃ³n en estadÃ­stica que representa la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1)
#de un conjunto de datos. Es una medida robusta que muestra la variabilidad del 50% central de los datos y es Ãºtil para identificar valores atÃ­picos, es decir
#puntos de datos que se encuentran muy alejados del resto del conjunto.
```

#4.3  Aplicar la tÃ©cnica de anÃ¡lisis

**Vamos a analizar la volatilidad de las empresas del anio 2022 en adelante.**

La **volatilidad** se define como la **desviaciÃ³n estÃ¡ndar de sus rendimientos (daily return)**, que mide la dispersiÃ³n de los precios respecto a la media a lo largo del tiempo.

Primero vamos a segmentar a las empresas de acuerdo a su volatilidad
1. Calcular la volatilidad de cada una de las empresas
2. Clasificar todas las empresas dependiendo de su volatilidad

Despues vamos a buscar alguna correlacion entre empresas que quedaron en segmentos similares.

3. Buscar si existe una correlacion entre las empresas que tienen mayor volatilidad y las que tienen menor volatilidad

##4.3.1 Aplicar segmentacion

###Crear un df con datos correspondientes al anio 2022 en adelante

```python
#Crear un df con datos correspondientes al anio 2022 en adelante
df2022 = df_prices_cleaned[df_prices_cleaned['date']>'2022']
print(df2022)
```

###Calcular la volatilidad del 2022 en adelante para cada una de las empresas y guardarlas en un DF:

```python
#Calcular la volatilidad del 2022 en adelante para cada una de las empresas y guardarlas en un DF:
def calcular_volatilidades(df, stock_symbols):
    """
    Calcula la volatilidad (desviaciÃ³n estÃ¡ndar del daily_return)
    de varias acciones y devuelve un DataFrame con los resultados.

    ParÃ¡metros:
        df (pd.DataFrame): DataFrame con columnas 'stock_symbol' y 'daily_return'.
        stock_symbols (list): Lista de sÃ­mbolos de acciones (ej. ['AAPL', 'TSLA']).

    Retorna:
        pd.DataFrame: con columnas ['stock_symbol', 'volatilidad'].
    """
    resultados = []
    for symbol in stock_symbols:
        datos = df[df['stock_symbol'] == symbol]
        vol = datos['daily_return'].std()
        resultados.append({'stock_symbol': symbol, 'volatilidad_2022+': vol})

    return pd.DataFrame(resultados)

# Ejemplo de uso:
stocks = ['AAPL','ADBE','AMZN', 'CRM','CSCO','GOOGL','IBM','INTC','META','MSFT','NFLX','NVDA','ORCL','TSLA']
df_volatilidades = calcular_volatilidades(df2022, stocks)

print(df_volatilidades)
```

###Clasificar cada empresa en cuartiles de volatilidad usando los valores de referencia (Q1, Q2, Q3).

```python
#Calcular cuartiles de la volatilidad:
Q1 = df_volatilidades['volatilidad_2022+'].quantile(0.25)
Q2 = df_volatilidades['volatilidad_2022+'].quantile(0.5)
Q3 = df_volatilidades['volatilidad_2022+'].quantile(0.75)

print(Q1)
print(Q2)
print(Q3)
```

```python
#Clasificar cada empresa en cuartiles de volatilidad usando los valores de referencia (Q1, Q2, Q3).

def clasificar_volatilidad(df):
    """
    Clasifica cada stock_symbol en cuartiles de volatilidad segÃºn Q1, Q2 y Q3.

    ParÃ¡metros:
        df (pd.DataFrame): DataFrame con columnas ['stock_symbol', 'volatilidad_2022+'].

    Retorna:
        pd.DataFrame: DataFrame con una nueva columna 'categoria_volatilidad'.
    """
        # Calcular cuartiles (A pesar de que se calcularon en el paso anterior, de esta manera quitas trabajo manual)
    Q1 = df["volatilidad_2022+"].quantile(0.25)
    Q2 = df["volatilidad_2022+"].quantile(0.50)  # mediana
    Q3 = df["volatilidad_2022+"].quantile(0.75)


    # Crear etiquetas
    etiquetas = ["Baja", "Media-Baja", "Media-Alta", "Alta"]

    # ClasificaciÃ³n usando pd.cut
    df["categoria_volatilidad"] = pd.cut(
        df["volatilidad_2022+"],
        bins=[-float("inf"), Q1, Q2, Q3, float("inf")],
        labels=etiquetas
    )

    return df

# Resultado:
df_clasificado = clasificar_volatilidad(df_volatilidades)
print(df_clasificado)
```

###Grafico de barras con la clasificacion por categoria de volatilidad 2022+ de cada empresa

```python
def graficar_barras_con_valores(df):
    """
    Crea un grÃ¡fico de barras con la volatilidad de cada empresa
    y muestra el valor de la volatilidad encima de cada barra.
    """
    plt.figure(figsize=(12,6))

    # Ordenar por volatilidad
    df_sorted = df.sort_values("volatilidad_2022+", ascending=True)

    # GrÃ¡fico de barras
    ax = sns.barplot(
        data=df_sorted,
        x="stock_symbol",
        y="volatilidad_2022+",
        hue="categoria_volatilidad",
        dodge=False,
        palette="viridis"
    )

    # Agregar etiquetas (solo valores)
    for i, row in enumerate(df_sorted.itertuples()):
        ax.text(
            i,
            row._2 + 0.05,  # un poco arriba de la barra (._2 es volatilidad_2022+)
            f"{row._2:.2f}",
            ha="center", va="bottom", fontsize=9
        )

    plt.title("Volatilidad de acciones (clasificadas por categorÃ­a)", fontsize=14)
    plt.xlabel("Empresa (stock_symbol)")
    plt.ylabel("Volatilidad 2022+")
    plt.legend(title="CategorÃ­a de volatilidad")
    plt.tight_layout()
    plt.show()

# Uso:
graficar_barras_con_valores(df_clasificado)

```

###(Borrador)Otras graficas para visualizar la distribucion de volatilidad 2022 por categoria

```python
def graficar_boxplot(df):
    """
    Genera un boxplot para visualizar la distribuciÃ³n de volatilidades por categorÃ­a.
    """
    plt.figure(figsize=(8,5))
    sns.boxplot(data=df, x="categoria_volatilidad", y="volatilidad_2022+",
                order=["Baja", "Media-Baja", "Media-Alta", "Alta"])

    plt.title("Volatilidad por categorÃ­a", fontsize=14)
    plt.xlabel("CategorÃ­a de volatilidad")
    plt.ylabel("Volatilidad 2022+")
    plt.show()

# Uso:
graficar_boxplot(df_clasificado)
```

```python

def graficar_categorias(df):
    """
    Genera un grÃ¡fico de barras con el nÃºmero de acciones por categorÃ­a de volatilidad.
    """
    plt.figure(figsize=(8,5))
    sns.countplot(data=df, x="categoria_volatilidad", order=["Baja", "Media-Baja", "Media-Alta", "Alta"])

    plt.title("DistribuciÃ³n de acciones segÃºn su volatilidad", fontsize=14)
    plt.xlabel("CategorÃ­a de volatilidad")
    plt.ylabel("NÃºmero de acciones")
    plt.show()

# Uso:
graficar_categorias(df_clasificado)
```

```python

def graficar_categorias_con_etiquetas(df):
    """
    Grafica las acciones segÃºn su categorÃ­a de volatilidad
    y muestra la etiqueta de cada stock_symbol.
    """
    plt.figure(figsize=(10,6))

    # Scatterplot de cada empresa
    sns.stripplot(
        data=df,
        x="categoria_volatilidad",
        y="volatilidad_2022+",
        order=["Baja", "Media-Baja", "Media-Alta", "Alta"],
        jitter=True,
        size=8
    )

    # Agregar etiquetas
    for i, row in df.iterrows():
        plt.text(
            x=["Baja", "Media-Baja", "Media-Alta", "Alta"].index(row["categoria_volatilidad"]),
            y=row["volatilidad_2022+"],
            s=row["stock_symbol"],
            fontsize=9,
            ha="center",
            va="bottom"
        )

    plt.title("Volatilidad de acciones por categorÃ­a (con etiquetas)", fontsize=14)
    plt.xlabel("CategorÃ­a de volatilidad")
    plt.ylabel("Volatilidad 2022+")
    plt.show()

# Uso:
graficar_categorias_con_etiquetas(df_clasificado)
```

##4.3.2 Aplicar Correlacion

###Correlacion entre metricas de empresas de volatilidad baja

```python

# 1ï¸ Filtrar empresas con volatilidad baja
empresas_baja = df_clasificado[df_clasificado["categoria_volatilidad"] == "Baja"]["stock_symbol"]

# 2ï¸ Filtrar el DataFrame diario para esas empresas
df_baja = df2022[df2022["stock_symbol"].isin(empresas_baja)]

# 3ï¸ Calcular mÃ©tricas promedio por empresa
df_resumen = df_baja.groupby("stock_symbol").agg({
    "adj_close": "mean",
    "volume": "mean",
    "daily_return": "mean"
}).reset_index()

# 4ï¸ Agregar la volatilidad para referencia (opcional)
df_resumen = df_resumen.merge(df_clasificado[["stock_symbol", "volatilidad_2022+"]], on="stock_symbol")
print(df_resumen)

# 5ï¸ Calcular matriz de correlaciÃ³n
corr = df_resumen[["adj_close", "volume", "daily_return", "volatilidad_2022+"]].corr()

# 6ï¸ Visualizar con heatmap
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap="Blues", fmt=".2f")
plt.title("CorrelaciÃ³n entre mÃ©tricas (empresas volatilidad baja)")
plt.show()
```

###Correlacion entre metricas de empresas de volatilidad alta

```python
#print('df_clasificado:', df_clasificado)
# 1ï¸ Filtrar empresas con volatilidad alta
empresas_alta = df_clasificado[df_clasificado["categoria_volatilidad"] == "Alta"]["stock_symbol"]
#print(empresas_alta)

# 2ï¸ Filtrar el DataFrame diario para esas empresas
df_alta = df2022[df2022["stock_symbol"].isin(empresas_alta)]#.isin() Se usa para filtrar datos en estructuras de Pandas (Series o DataFrames) comprobando si los elementos de la estructura se encuentran dentro de una lista, otra Serie o un DataFrame
#print(df_alta)

# 3ï¸ Calcular mÃ©tricas promedio por empresa
df_resumen = df_alta.groupby("stock_symbol").agg({
    "adj_close": "mean",
    "volume": "mean",
    "daily_return": "mean"
}).reset_index()
#print('df_resumen: ',df_resumen)

# 4ï¸ Agregar la volatilidad para referencia (opcional)
df_resumen = df_resumen.merge(df_clasificado[["stock_symbol", "volatilidad_2022+"]], on="stock_symbol")
print(df_resumen)

# 5ï¸ Calcular matriz de correlaciÃ³n
corr = df_resumen[["adj_close", "volume", "daily_return", "volatilidad_2022+"]].corr()

# 6ï¸ Visualizar con heatmap
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap="Blues", fmt=".2f")
plt.title("CorrelaciÃ³n entre mÃ©tricas (empresas volatilidad alta)")
plt.show()
```
